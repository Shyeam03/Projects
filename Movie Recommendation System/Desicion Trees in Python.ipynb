```python
import pandas as pd
```

```python
movies=pd.read_csv("Movie_classification.csv")
movies.head(10)
```

```python
# Discretizing the data
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
Genre_n=le.fit_transform(movies['Genre'])
```

```python
d=[]
a=-1
for i in movies['3D_available']:
    if i=="YES":
        a=1
        d.append(a)
    elif i=="NO":
        a=0
        d.append(a)
movies['3d_available']=d
```

```python
movies['Genre_n']=Genre_n.tolist()
```

```python
movies.drop(labels=['3D_available','Genre','Time_taken'],axis=1,inplace=True)
```

```python
movies
```

```python
#Take out the label Collection
#We are trying to predict collection
X=movies.loc[:,movies.columns!='Collection']
y=movies['Collection']
```

```python
#Test Train Split
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25)
```

```python
#Traning Decision Tree
from sklearn import tree
regtree=tree.DecisionTreeRegressor(max_depth=3)
regtree.fit(X_train,y_train)
```

```python
#Preciting values
y_train_pred=regtree.predict(X_train)
y_test_pred=regtree.predict(X_test)
```

```python
import numpy as np
a=np.around(y_test_pred)
a.astype(int)
print(a)
```

```python
from sklearn.metrics import mean_squared_error,r2_score
mean_squared_error(y_test,y_test_pred)
```

```python
import sklearn.metrics as metrics
print('Root Mean Squared Error for test data:', np.sqrt(metrics.mean_squared_error(y_test, y_test_pred)))
```

```python
print('Root Mean Squared Error for train data:', np.sqrt(metrics.mean_squared_error(y_train, y_train_pred)))
```

```python
r2_score(y_train,y_train_pred)
# the value obtained is 0.83 which means our model is performing great
```

```python
r2_score(y_test,y_test_pred)
```

```python
#Confusion matrix
import seaborn as sns
cm=confusion_matrix(y_test,a)
print(cm)
plt.figure(figsize = (14,16))
sns.heatmap(cm,annot=True)
```

```python
#Plotting the trees
dot_data=tree.export_graphviz(regtree, out_file=None)
from IPython.display import Image
import pydotplus
graph=pydotplus.graph_from_dot_data(dot_data)

Image(graph.create_png())
```
